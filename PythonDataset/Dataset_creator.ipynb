```python
import os
import json
```

```python
python_files = []

for root, directories, files in os.walk("."):
    for file in files:
        if file.endswith(".py"):
            if '_cwe' in file and 'test_' not in file:
                python_files.append(os.path.join(root, file))

print(len(python_files))
```

```python
dataset = []
# Print the list of Python files
for file in python_files:

    id = "_".join(file.split('/')[1:])
    print(id)
    technique = id.split('_')[0]
    print(technique)
    source = id.split('_')[1]
    print(source)
    with open(file) as f:
        insecure_code = f.read()
    prompt = "'''".join(insecure_code.split("'''")[:-1]) + "'''"
    # test_code = ''  
    # if technique == 'Assertion':
    #     # add 'test' in the file name after second '/'
    #     test_file = file.split('/')
    #     test_file[3] = 'test_'+test_file[3]
    #     test_file = '/'.join(test_file)
    #     with open(test_file) as f:
    #         test_code = f.read()
    

    dataset.append({
        'id': id,
        'technique': technique,
        'source': source,
        'prompt': prompt,
        'insecure_code': insecure_code,
        # 'test_code': test_code
    })
```

```python
len(dataset)
```

```python
with open('dataset.jsonl', 'w') as f:
    for data in dataset:
        json.dump(data, f)
        f.write('\n')
```
